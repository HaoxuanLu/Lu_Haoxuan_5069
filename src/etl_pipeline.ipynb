```python
df_laptimes = spark.read.csv("/Volumes/gr5069/raw/f1_data/lap_times.csv", header = True)
```

```python
display(df_laptimes)
```

```python
df_drivers = spark.read.csv("/Volumes/gr5069/raw/f1_data/drivers.csv", header = True)
display(df_drivers)
```

## Transform Data for Drivers and added 'age' 

```python
# calculate the age of all the drivers 
df_drivers = df_drivers.withColumn('age', floor(datediff(current_date(), col('dob'))/365))
display(df_drivers)
```

```python
df_lap_drivers = df_laptimes.join(df_drivers, df_laptimes.driverId == df_drivers.driverId)
display(df_lap_drivers)
```

```python
df_lap_drivers = df_lap_drivers.groupby('age').agg(avg('lap').alias('avg_lap'))
display(df_lap_drivers)
```

```python
df_lap_drivers.write.csv("/Volumes/gr5069/ne2388/inclass/hl3921_lap_drivers_output.csv", header = True)
```
